#### Import Package

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
```

#### Load Model

```python
# Load model
model_name = "cahya/bert-base-indonesian-NER"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Setup NER pipeline
ner_pipeline = pipeline(
    "ner",
    model=model,
    tokenizer=tokenizer,
    aggregation_strategy="simple"  # Group tokens into entities
)
```

#### Load Data

```python
kompas_data = pd.read_csv("data/kompas.csv")
kompas_data.head()
```

```python
detik_data = pd.read_csv("data/detik.csv")
detik_data.head()
```

```python
# Concate data
df = pd.concat([kompas_data, detik_data], ignore_index=True)
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# Get only 1 row for testing
row = df.iloc[0]

# Concate the title and description to text for NER
text = row['title']
print(text)
```

#### Apply NER

```python
# Apply NER
entities = ner_pipeline(text)

# See the results
for ent in entities:
    print(f"{ent['word']} -> {ent['entity_group']} ({ent['score']:.2f})")
```

```python
df_results = []

# Apply in all row data
for i, row in df.iterrows():
    text = f"{row['title']}"
    entities = ner_pipeline(text)
    for ent in entities:
        df_results.append({
            'article_id': i,
            'entity': ent['word'],
            'label': ent['entity_group'],
            'score': ent['score']
        })

# Save results
df_ner = pd.DataFrame(df_results)
df_ner.to_csv("ner_results.csv", index=False)
```
